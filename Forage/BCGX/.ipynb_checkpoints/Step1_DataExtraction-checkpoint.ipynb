```python
import pandas as pd
```

```python
df = pd.read_csv('10-k.csv')
```

```python
display(df)
```

```python
df = df.dropna()
```

```python
display(df)
```

```python
print(df.dtypes)
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change()*100
```

```python
df['Net Income Growth(%)'] = df.groupby(['Company'])['Net Income'].pct_change()*100
```

```python
display(df)
```

```python
df.fillna(0, inplace=True)
```

```python
display(df)
```

```python
# Aggregate metrics by company
company_agg = df.groupby("Company").agg({
    "Total Revenue": ["sum", "mean", "max"],
    "Net Income": ["sum", "mean", "max"],
    "Total Assets": ["sum", "mean", "max"],
    "Total Liabilities": ["sum", "mean", "max"],
    "Cash Flow from Operating Activities": ["sum", "mean", "max"]
}).reset_index()
```

```python
# Aggregate metrics by year
year_agg = df.groupby("Year").agg({
    "Total Revenue": ["sum", "mean", "max"],
    "Net Income": ["sum", "mean", "max"],
    "Total Assets": ["sum", "mean", "max"],
    "Total Liabilities": ["sum", "mean", "max"],
    "Cash Flow from Operating Activities": ["sum", "mean", "max"]
}).reset_index()
```

```python
# Print the aggregated data
print("Aggregated Data by Company:")
display(company_agg)
```

```python
print("\nAggregated Data by Year:")
display(year_agg)
```

# **Financial Performance Analysis Summary**

### **1. Revenue & Net Income Trends**
- Microsoft, Tesla, and Apple have shown **consistent revenue growth** over the years.
- **Apple** has the highest **total revenue**, with its peak in 2024.
- **Tesla** has lower revenue compared to Microsoft and Apple but has shown a strong **net income growth rate**.

### **2. Assets & Liabilities Insights**
- Microsoft has **the highest total assets**, indicating strong financial backing.
- Tesla has a lower asset base but also maintains **relatively lower liabilities**.
- Liabilities have **increased across all companies** in 2024, which may indicate higher investments or financial obligations.

### **3. Cash Flow from Operating Activities**
- Apple and Microsoft maintain **strong operating cash flow**, reinforcing their ability to generate cash from core business operations.
- Tesla, despite growing revenue, has significantly lower cash flow, suggesting **higher reinvestment costs or lower operational efficiency**.

### **4. Yearly Comparisons**
- Total revenue and net income **increased across all three companies in 2024**, compared to 2022 and 2023.
- 2024 recorded **the highest total revenue** across companies, showing strong market performance.
